Найдем кратчайший маршрут от вершины 0 до вершины 13 с помощью алгоритма Дейкстры.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
for i in range(14):
    G.add_node(i)

# первая строчка
G.add_edge(0, 1, weight=6)
G.add_edge(0, 4, weight=4)
G.add_edge(0, 2, weight=3)
G.add_edge(1, 3, weight=2)
G.add_edge(1, 6, weight=2)
G.add_edge(1, 4, weight=1)
# вторая строчка
G.add_edge(2, 4, weight=2)
G.add_edge(2, 7, weight=10)
G.add_edge(2, 5, weight=6)
G.add_edge(3, 8, weight=3)
G.add_edge(3, 6, weight=1)
G.add_edge(4, 6, weight=4)
# третья строчка
G.add_edge(4, 9, weight=7)
G.add_edge(6, 7, weight=6)
G.add_edge(4, 7, weight=9)
G.add_edge(5, 7, weight=4)
G.add_edge(5, 10, weight=3)
G.add_edge(6, 8, weight=4)
# четвертая строчка
G.add_edge(6, 11, weight=8)
G.add_edge(6, 9, weight=3)
G.add_edge(7, 9, weight=2)
G.add_edge(7, 12, weight=6)
G.add_edge(7, 10, weight=2)
G.add_edge(8, 11, weight=7)
# пятая строчка
G.add_edge(9, 11, weight=6)
G.add_edge(9, 13, weight=13)
G.add_edge(9, 12, weight=9)
G.add_edge(10, 12, weight=7)
G.add_edge(11, 13, weight=8)
G.add_edge(12, 13, weight=4)

pos = nx.spring_layout(G)
nx.draw(G, pos)
nx.draw_networkx_nodes(G, pos, node_color='black')
nx.draw_networkx_nodes(nx.dijkstra_path(G, 0, 13), pos, node_color='blue')
nx.draw_networkx_labels(G, pos, font_color='white')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='black')
print("Кратчайший путь - по узлам:", nx.dijkstra_path(G, 0, 13))
plt.show()
```
